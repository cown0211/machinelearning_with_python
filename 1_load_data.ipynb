```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
raw_boston = datasets.load_boston() # sklearn으로부터 데이터 로드
X_boston = pd.DataFrame(raw_boston.data) # X 데이터
y_boston = pd.DataFrame(raw_boston.target) # y 타겟
df_boston = pd.concat([X_boston, y_boston], axis =  1) # concat 함수로 df 묶어줌, axis = 0; 위+아래, axis = 1; 좌+우
```

```python
df_boston
```

```python
df_boston.info()
```

```python
feature_boston = raw_boston.feature_names
print(feature_boston)
# 변수 이름
```

```python
col_boston = np.append(feature_boston, ['target'])
df_boston.columns = col_boston
df_boston.head()
```
